# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
engine.execute('SELECT * FROM Measurement LIMIT 5').fetchall()
```

```python
session.query(Measurement.date).\
    filter(Measurement.date > '2018-03-01').\
    order_by(Measurement.date).all()
```

```python
# Calculate the date 1 year ago from the last data point in the database
latest_date = session.query(func.max(Measurement.date)).one()[0]
from datetime import datetime

latest_date = datetime.strptime(latest_date, '%Y-%m-%d')
latest_date
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

yearago = latest_date
year = latest_date.year - 1
day = latest_date.day - 1
yearago = yearago.replace(year = year)
yearago = yearago.replace(day = day)
print(f"Query time filter: {yearago}")
# Perform a query to retrieve the data and precipitation scores
prcp_year = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date > yearago).all()

# Save the query results as a Pandas DataFrame and set the index to the date column
df = pd.DataFrame(prcp_year, columns=['Date', 'Precipitation'])

# Sort the dataframe by date
df.sort_values(by=['Date'])
df.set_index('Date', inplace=True)
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
df.plot(figsize = (12,10),title = "Precipitation for latest 12 months" )
plt.tick_params(axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) 
plt.ylabel("Precipitation (in mm)")
plt.tight_layout()
plt.savefig("Precipitation.png")
```

```python
# Design a query to show how many stations are available in this dataset?
session.query(func.count(Station.station)).all()
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
active_prcp = session.query(Measurement.station,func.count(Measurement.prcp),).\
group_by(Measurement.station).order_by(func.count(Measurement.prcp)).all()

most_active_station_prcp = active_prcp[-1][0]
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?
session.query(Measurement.station,func.min(Measurement.tobs),func.max(Measurement.tobs),func.avg(Measurement.tobs)).\
group_by(Measurement.station).filter(Measurement.station == most_active_station_prcp).all()
```

```python
# Choose the station with the highest number of temperature observations.
active_tobs = session.query(Measurement.station,func.count(Measurement.tobs)).\
group_by(Measurement.station).order_by(func.count(Measurement.tobs)).all()

most_active_station_tobs = active_tobs[-1][0]

# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
temp_year = session.query(Measurement.date, Measurement.tobs).filter(Measurement.date > yearago).\
filter(Measurement.station == most_active_station_tobs).all()

# Save the query results as a Pandas DataFrame and set the index to the date column
df_temp = pd.DataFrame(temp_year, columns=['Date', 'Temperature'])

# Sort the dataframe by date
df_temp.sort_values(by=['Date'])
df_temp.set_index('Date', inplace=True)
df_temp.head()

# plot histogram
df_temp.plot(kind = 'hist', figsize = (12,10),title = "Temperature in latest 12 months", xticks = None, bins=12)
plt.tick_params(axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) 
plt.xlabel("Dates")
plt.tight_layout()
plt.savefig("Temperature_year.png")
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
results = (calc_temps('2012-02-28', '2012-03-05'))
temp = list(np.ravel(results))
temp
```

```python
#Calculate the range for last one year data
result = (calc_temps(yearago, latest_date))
temp_data = list(np.ravel(result))

#convert to a dataframe
year_temp = pd.DataFrame([[temp_data[0], temp_data[1], temp_data[2]]],
                       columns=['Min Temp', 'Average Temp', 'Max Temp'])

# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
year_temp['Spread'] = year_temp['Max Temp'] - year_temp['Min Temp']

# Plot the results from your previous query as a bar chart.
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
year_temp.plot.bar(y = 'Average Temp', title = "Trip Avg Temp", yerr = 'Spread')

#Removing the ticks and bunch of values from x
plt.tick_params(axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) 

#label y axis
plt.ylabel("Temperature (F)")
plt.tight_layout()
plt.savefig("TripAvgTemp.png")
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```
